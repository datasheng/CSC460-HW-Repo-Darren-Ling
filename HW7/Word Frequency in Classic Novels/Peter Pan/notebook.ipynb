```python
# Use this cell to begin your analysis, and add as many as you would like!

# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import nltk
from collections import Counter
```

```python
# Create html request
r = requests.get("https://www.gutenberg.org/files/16/16-h/16-h.htm")
r.encoding = 'utf-8'
html = r.text
```

```python
# Create BeautifulSoup and Tokens
soup = BeautifulSoup(html)
text = soup.get_text()

tokenizer = nltk.tokenize.RegexpTokenizer('\w+')
tokens = tokenizer.tokenize(text)
```

```python
# Make all words lowercase
words = [token.lower() for token in tokens]

# load stop words
sw = nltk.corpus.stopwords.words('english')

# Remove stop words
words_ns = [word for word in words if word not in sw]
```

```python
# Create a counter for list of words
counter = Counter(words_ns)
top_ten = counter.most_common(10)
print(top_ten)

# Most common name
protagonists = ["peter", "wendy", "hook", "john"]
```
